```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\CVR\\Downloads\\archive(1)\\train.csv')
sns.barplot(data=df,x='Pclass',y='Survived',errorbar=None)
plt.title('survival rate by passenger class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival rate')
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
sns.displot(penguins, x="flipper_length_mm", hue="species", kde=True)
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived',palette='coolwarm',alpha=0.6)
plt.title("Age vs Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
subset=df[['Age','Fare','Pclass','Survived']]
sns.pairplot(subset,hue='Survived',palette='coolwarm',diag_kind='kde')
plt.suptitle('Pairwise Relationships',y=1.02)
plt.show()
```

```python
num_cols=df.select_dtypes(include='number')
corr_matrix=num_cols.corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title("correlation matrix heatmap")
plt.show()
```
