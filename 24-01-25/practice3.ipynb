```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.title("Simple Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.plot(x,y,color="brown",linestyle="dotted",linewidth=2,marker='*')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[500,1000,4000,6000,8000,10000]
plt.title("Simple Line Plot")
plt.xlabel("Months(1-6)")
plt.ylabel("Saving Amount")
plt.grid()
plt.plot(x,y,color="darkgreen",linestyle="dashed",linewidth=2,marker='*')
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['royalblue','red','green','yellow','hotpink'],edgecolor='black')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20,10,30,25,15]
plt.pie(values, labels=categories, colors=['royalblue', 'red', 'green', 'yellow', 'hotpink'],autopct='%1.1f%%')
plt.legend()
plt.title("Basic Pie Chart")
plt.show()
```
