```python
'''Display all the records whose start city is Newyork or cary or Daytona beach and stop city is West palm beach 
and jamaica and morrisville and miles between 10 to 20'''

filtered_data = uber_data.loc[
    (uber_data['START*'].isin(['New York', 'Cary', 'Daytona Beach'])) &
    (uber_data['STOP*'].isin(['West Palm Beach','Jamaica', 'Morrisville'])) &
    (uber_data['MILES*'] >= 10) &
    (uber_data['MILES*'] <= 20) 
]
filtered_data
```

```python
'''Display all the records of Jan-2016'''

# Convert 'START_DATE*' to datetime format
uber_data['START_DATE*'] = pd.to_datetime(uber_data['START_DATE*'])

# Drop rows where 'START_DATE*' could not be converted (NaT values)
uber_data = uber_data.dropna(subset=['START_DATE*'])

# Now filter the data to include only dates within January 2016
filtered_data = uber_data.loc[
    (uber_data['START_DATE*'] >= '2016-01-01') & 
    (uber_data['START_DATE*'] <= '2016-01-31')
]

filtered_data
```
